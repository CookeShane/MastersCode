#### Imports

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

***
#### Plot Training vs Validation Loss

```python
train = pd.read_csv('train_loss.csv')
val = pd.read_csv('val_loss.csv')

train_loss_x_values = train['Step']
train_loss_y_values = train['Value']

val_loss_x_values = val['Step']
val_loss_y_values = val['Value']
```

```python
plt.plot(train_loss_x_values, train_loss_y_values, label='Training Loss')
plt.plot(val_loss_x_values, val_loss_y_values, label='Validation Loss')

plt.title('Training vs Validation Loss (Adam)', fontsize=18)
plt.xlabel('Epoch', fontsize=16)
plt.ylabel('Loss', fontsize=16)

x_labels = list(range(1, (len(val_loss_x_values)+1)))
plt.xlim([0, max(val_loss_x_values)])
plt.ylim([0,1])
plt.xticks(ticks=val_loss_x_values, labels=x_labels, fontsize=12)
plt.yticks(ticks=np.arange(0,1.1,0.1), fontsize=12)
plt.grid(axis='x')
plt.gcf().set_size_inches(15,8)
plt.legend(fontsize=16, loc='upper right')

plt.savefig('Training_Validation_Loss.jpg', dpi=1000)
plt.show()
```
